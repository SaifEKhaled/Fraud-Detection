```python
df_isfraud_0 = df[df['isFraud'] == 0].head(8000)

df_isfraud_1 = df[df['isFraud'] == 1].head(8000)

new_dataset = pd.concat([df_isfraud_0, df_isfraud_1], ignore_index=True)

```

```python
new_datasetCorr = new_dataset[['step','type','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']].corr()
sns.heatmap(new_datasetCorr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of the new dataset')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
col = ["step", "type", "amount", "oldbalanceOrg", "newbalanceOrig"]
x = np.array(new_dataset[col])
y = np.array(new_dataset["isFraud"])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
model1=DecisionTreeClassifier()
model2=LogisticRegression()
model3=RandomForestClassifier()
model4=KNeighborsClassifier()
model1.fit(x_train,y_train.ravel())
model2.fit(x_train,y_train.ravel())
model3.fit(x_train,y_train.ravel())
model4.fit(x_train,y_train.ravel())
```

```python
from sklearn.svm import SVC
model5 = SVC(probability=True) 
model5.fit(x_train,y_train)
```

```python
predictions1 = model1.predict(x_test)
predictions2 = model2.predict(x_test)
predictions3 = model3.predict(x_test)
predictions4 = model4.predict(x_test)
```

```python
predictions5=model5.predict(x_test)
```

## Confusion Metrics

```python
from sklearn.metrics import confusion_matrix,accuracy_score


cm = confusion_matrix(y_test, predictions1)


sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)

accuracy = accuracy_score(y_test, predictions1)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Model1(DecisionTreeClassifier)\nAccuracy: {accuracy:.4f}")
plt.show()
```

```python
cm = confusion_matrix(y_test, predictions2)


sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
accuracy = accuracy_score(y_test, predictions2)

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Model2(LogisticRegression)\nAccuracy: {accuracy:.4f}")
plt.show()
```

```python
cm = confusion_matrix(y_test, predictions3)


sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
accuracy = accuracy_score(y_test, predictions3)

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Model3(RandomForestClassifier)\nAccuracy: {accuracy:.4f}")
plt.show()
```

```python
cm = confusion_matrix(y_test, predictions4)


sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
accuracy = accuracy_score(y_test, predictions4)

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Model4(KNeighborsClassifier)\nAccuracy: {accuracy:.4f}")
plt.show()
```

```python
cm = confusion_matrix(y_test, predictions5)


sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
accuracy = accuracy_score(y_test, predictions5)

plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Model5(SVC)\nAccuracy: {accuracy:.4f}")
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier
voting_classifier = VotingClassifier(
    estimators=[('dt', model1), ('lr', model2), ('rf', model3), ('knn', model4), ('svc', model5)],
    voting='hard'  
)

voting_classifier.fit(x_train, y_train)

predictions_voting = voting_classifier.predict(x_test)

accuracy_voting = accuracy_score(y_test, predictions_voting)
print(f"Accuracy of the ensemble model: {accuracy_voting:.4f}")

cm_voting = confusion_matrix(y_test, predictions_voting)
sns.heatmap(cm_voting, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title(f"Confusion Matrix for Ensemble Model\nAccuracy: {accuracy_voting:.4f}")
plt.show()
```
